# Differential Field Analysis via PCA

This notebook demonstrates how to analyze the "differential field" created by a semantic intervention (e.g., adding an identity vector).

We use PCA to find the principal directions in which the model's score function changes in response to the intervention across many points in latent space.

```python
from pathlib import Path
import sys
import numpy as np
import matplotlib.pyplot as plt

root = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.insert(0, str(root))

from client.config import registry
from client.interface import SlopClient
from client.utils.identity import extract_identity_vector

providers = registry.list()
client = SlopClient(providers[-1])
client.connect()
print(f'Connected to {client.info().gpu_name}')
```

```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from client.utils.identity import extract_identity_vector, apply_identity
```

```python

# ============================================================
# extract the identity vector
# ============================================================

# isolate "Arab" direction by contrasting across multiple contexts
# this is more robust than single prompt pairs
activities = [
    "shopping in a city",
    "sitting on the beach", 
    "eating fruit",
    "playing videogames",
    "relaxing",
    "praying"
]

print("extracting identity vector...")
arab_vector = extract_identity_vector(client, "Arab", activities)
print(f"identity vector shape: {arab_vector.shape}")
```

```python
base_prompt = "person in a city"

# get base embeddings
base_embeds, _ = client.embed(base_prompt)
target_embeds = apply_identity(base_embeds, arab_vector, scale=5.0)
```

```python
# create stable point to make sure th field e got lie on the stable manifold parts
from shared.physics.types import Field
from shared.physics.operators import divergence as div
from shared.physics.latent import LatentField, LatentDiffField


f = LatentField(client, prompt_embeds=base_embeds, negative_prompt_embeds=None, max_batch=64, flush_ms=1.0)
d = LatentDiffField(client, target_embeds=target_embeds, base_embeds=base_embeds, negative_prompt_embeds=None, max_batch=64, flush_ms=1.0)
# find stable points on the image manifold

# find n stable points
n_stable = 1
stable_points = []
max_steps = 100
step_size = 0.1

print(f"finding {n_stable} stable points...\n")

for i in range(1,n_stable+1):
    # random initialization
    current = np.random.randn(1, 4, 64, 64).astype(np.float32)
    converged = False
    
    print(f"stable point {i+1}/{n_stable}:")
    for step in range(max_steps):
        force = f(current)
        divergence = div(f, current)
        
        if step % 10 == 0:
            print(f"  step {step}: div={divergence:.4f}")
        
        # check stability
        if abs(divergence) < 0:
            print(f"  converged at step {step}")
            converged=True
            break
        
        # integrate
        current = current + (step_size/(i/10)) * force
    

    stable_points.append(stable)
    print()

for s in stable_points:
    r = client.render(s)
    plt.figure(figsize=(8, 8))
    plt.imshow(r[0])
    plt.axis('off')
    plt.title('Generated Image')
    plt.show()

stable_points = np.array(stable_points)  # shape: (n_stable, 1, 4, 64, 64)
print(f"found {len(stable_points)} stable points")
```

```python

# ============================================================
# sample differential field near stable points
# ============================================================

# for each stable point, sample neighborhood
n_samples_per_point = 10
neighborhood_radius = 0.5

all_samples = []
all_diffs = []

print("\nsampling differential field near stable points...")

for i, stable in enumerate(stable_points):
    # sample points in neighborhood
    perturbations = np.random.randn(n_samples_per_point, 4, 64, 64) * neighborhood_radius
    neighborhood = stable + perturbations.astype(np.float32)
    
    # evaluate differential at each
    diffs = d(neighborhood)
    
    all_samples.append(neighborhood)
    all_diffs.append(diffs)
    
    if (i+1) % 10 == 0:
        print(f"  processed {i+1}/{len(stable_points)} stable points")

all_samples = np.concatenate(all_samples, axis=0)
all_diffs = np.concatenate(all_diffs, axis=0)

print(f"\ntotal samples: {len(all_samples)}")
print(f"differential field shape: {all_diffs.shape}")
```

```python
from math import ceil

timestep = 500  
n_points = 50
batch_size = 8


# sample random latent points
latents = np.random.randn(n_points, 4, 64, 64).astype(np.float32)

base_scores = []
target_scores = []

# batch processing
for i in range(ceil(n_points / batch_size)):
    start = i * batch_size
    end = min((i + 1) * batch_size, n_points)
    batch = latents[start:end]
    
    print(f"probing batch {i+1}/{ceil(n_points/batch_size)} ({end-start} points)...")
    base_batch = client.probe_at(batch, base_embeds, timestep=timestep)
    target_batch = client.probe_at(batch, target_embeds, timestep=timestep)
    
    base_scores.append(base_batch)
    target_scores.append(target_batch)

# concatenate all batches
base_scores = np.concatenate(base_scores, axis=0)
target_scores = np.concatenate(target_scores, axis=0)

# differential field
diff_field = target_scores - base_scores

print(f"\nsampled {n_points} points in latent space")
print(f"differential field shape: {diff_field.shape}")
```

```python
# ============================================================
# PCA: find dominant bias directions
# ============================================================

diff_flat = diff_field.reshape(n_points, -1)
diff_centered = diff_flat - diff_flat.mean(axis=0)

pca = PCA()
pca.fit(diff_centered)

variance_explained = np.cumsum(pca.explained_variance_ratio_)

print(f"\n=== BIAS CONCENTRATION ===")
print(f"top 1 PC:   {variance_explained[0]:.1%}")
print(f"top 5 PCs:  {variance_explained[4]:.1%}")


eff_dim = np.argmax(variance_explained > 0.9) + 1
print(f"\neffective dimensionality (90% variance): {eff_dim} / {diff_flat.shape[1]}")

if eff_dim < 50:
    print("  -> HIGHLY STEREOTYPED (narrow concept of identity)")
elif eff_dim < 200:
    print("  -> MODERATELY CONCENTRATED (structured bias)")
else:
    print("  -> DIFFUSE (either diverse or high-entropy)")
```

```python
# ============================================================
# magnitude analysis
# ============================================================

magnitude = np.linalg.norm(diff_flat, axis=1)

print(f"\n=== MAGNITUDE STATS ===")
print(f"mean:   {magnitude.mean():.4f}")
print(f"std:    {magnitude.std():.4f}")
print(f"range:  [{magnitude.min():.4f}, {magnitude.max():.4f}]")

cv = magnitude.std() / magnitude.mean()
print(f"CV:     {cv:.4f}")
print(f"  -> {'uniform across latent space' if cv < 0.3 else 'spatially varying (context-dependent)'}")
```

```python

# ============================================================
# visualize
# ============================================================

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# eigenvalue spectrum
axes[0].plot(pca.explained_variance_ratio_[:100])
axes[0].set_xlabel('principal component')
axes[0].set_ylabel('variance explained')
axes[0].set_title('eigenvalue spectrum\n(how concentrated is the bias?)')
axes[0].set_yscale('log')
axes[0].grid(True, alpha=0.3)

# cumulative variance
axes[1].plot(variance_explained[:200])
axes[1].axhline(0.9, color='red', linestyle='--', alpha=0.5, label='90%')
axes[1].axvline(eff_dim, color='red', linestyle='--', alpha=0.5)
axes[1].set_xlabel('number of components')
axes[1].set_ylabel('cumulative variance')
axes[1].set_title(f'cumulative variance\n(eff dim = {eff_dim})')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

# magnitude distribution
axes[2].hist(magnitude, bins=50, alpha=0.7, edgecolor='black')
axes[2].axvline(magnitude.mean(), color='red', linestyle='--', label='mean')
axes[2].set_xlabel('differential magnitude')
axes[2].set_ylabel('count')
axes[2].set_title('spatial distribution of bias')
axes[2].legend()
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python

# project each point's differential into PC space
pc_coordinates = pca.transform(diff_centered)  # shape: (n_points, 38)
def explore_pc(pc_idx, n_samples=5, denoise_steps=20):
    """
    find latents with extreme PC scores, denoise unconditionally, then render
    """
    scores = pc_coordinates[:, pc_idx]
    
    high_idx = np.argsort(scores)[-n_samples:]
    low_idx = np.argsort(scores)[:n_samples]
    
    print(f"\nPC{pc_idx + 1} (explains {pca.explained_variance_ratio_[pc_idx]:.1%})")
    print(f"score range: [{scores.min():.2f}, {scores.max():.2f}]")
    
    high_latents = latents[high_idx]
    low_latents = latents[low_idx]
    
    
    print(f"denoising high-scoring latents unconditionally ({denoise_steps} steps)...")
    high_denoised = []
    for lat in high_latents:
        result = client.sample_from_embeds(
            base_embeds, neg,
            num_steps=denoise_steps,
            batch_size=1,
            initial_latent=lat[None, ...]
        )
        high_denoised.append(result.points[-1][0])
    
    print(f"denoising low-scoring latents unconditionally ({denoise_steps} steps)...")
    low_denoised = []
    for lat in low_latents:
        result = client.sample_from_embeds(
            base_embeds, neg,
            num_steps=denoise_steps,
            batch_size=1,
            initial_latent=lat[None, ...]
        )
        low_denoised.append(result.points[-1][0])
    
    print("rendering...")
    high_images = [client.render(lat[None, ...])[0] for lat in high_denoised]
    low_images = [client.render(lat[None, ...])[0] for lat in low_denoised]
    
    return high_images, low_images, scores[high_idx], scores[low_idx]
# explore the top few PCs
for pc_idx in range(5):  # look at top 5
    high_imgs, low_imgs, high_scores, low_scores = explore_pc(pc_idx, n_samples=5)
    
    # visualize
    fig, axes = plt.subplots(2, 5, figsize=(15, 6))
    
    for i in range(5):
        axes[0, i].imshow(high_imgs[i])
        axes[0, i].axis('off')
        axes[0, i].set_title(f'+{high_scores[i]:.1f}', fontsize=10)
        
        axes[1, i].imshow(low_imgs[i])
        axes[1, i].axis('off')
        axes[1, i].set_title(f'{low_scores[i]:.1f}', fontsize=10)
    
    axes[0, 0].set_ylabel('high PC score', fontsize=12, rotation=0, ha='right', va='center')
    axes[1, 0].set_ylabel('low PC score', fontsize=12, rotation=0, ha='right', va='center')
    
    plt.suptitle(f'PC{pc_idx + 1} ', y=0.98)
    plt.tight_layout()
    plt.show()
```
