# Remote Training via SlopClient

Train a Stable Diffusion model on collected teacher images using the remote server.


## 1. Configuration

```python
# Paths on the REMOTE server
REMOTE_MANIFEST_PATH = "/root/slop/distill/data/test/grok_test/manifest.jsonl"
REMOTE_OUTPUT_DIR = "/root/slop/distill/output/grok_test"

# Training config
MODEL_ID = "runwayml/stable-diffusion-v1-5"
BATCH_SIZE = 2
EPOCHS = 1
LEARNING_RATE = 5e-5
SAVE_EVERY = 50
```

## 2. Connect to Remote Server

```python
from pathlib import Path
import sys
root = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.insert(0, str(root))

from client.config import registry
from client.interface import SlopClient


# Get provider
print(registry.list())
provider = registry.list()[-1]
print(f"Provider: {provider.name}")

# Connect
client = SlopClient(provider)
client.connect()

# Check server info
info = client.info()
print(f"Server: {info.hostname}")
print(f"GPU: {info.gpu_name}")
print(f"GPU Memory: {info.gpu_memory_mb}MB")
```

## 3. Run Remote Training

```python
from client.config import registry

# Force reload from disk (in case notebook was imported before providers existed)
registry.refresh()

PROVIDERS = registry.list()
print(f"Found {len(PROVIDERS)} providers: {[p.name for p in PROVIDERS]}")

if not PROVIDERS:
    raise RuntimeError("No providers found. Add a provider first.")

# Use the first provider, or set PROVIDER_NAME explicitly
provider = PROVIDERS[0]
print(f"Using provider: {provider.name}")
```

## 4. Cleanup

```python
# Close connection
client.close()
print("Connection closed")
```
