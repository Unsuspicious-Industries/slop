# Full Distillation Workflow

This notebook runs the end-to-end distillation collection pipeline on prompts
derived from the historical dataset. It collects teacher images, writes a
manifest, and shows a few sample images.


```python
# Setup and configuration
from distill.env import load_dotenv_if_present
from distill.config import DistillConfig, apply_model_preset
from distill.collect import collect
from distill.prompts import load_historical_prompts
from pathlib import Path
import json

load_dotenv_if_present()
cfg = DistillConfig(
    run_id='historical_full',
    dataset_name='historical',
    teacher='openrouter',
    model_preset='sd15',
    samples_per_prompt=2,
    max_concurrent=4,
    partial_supervision=False,
)
cfg = apply_model_preset(cfg)

# Load prompts from historical embeddings labels (cap to 200 prompts here)
prompts = load_historical_prompts('data/historical/embeddings', max_count=200)
print('Prompts loaded:', len(prompts))
```

## Start collection
The cell below runs the collection pipeline and writes the manifest and samples
to `distill/data/historical/<run_id>/` (default). This can take time and will use
your configured teacher API. Make sure `OPENROUTER_API_KEY` is set in `.env`.


```python
manifest = collect(cfg)
print('Manifest written to', manifest)
```

## Inspect a few samples
Read the manifest and display a handful of collected images inline.


```python
from IPython.display import display, Image
from distill.collect import MANIFEST_NAME
root = Path('distill/data/historical') / cfg.run_id
manifest_path = root / MANIFEST_NAME
with manifest_path.open('r', encoding='utf-8') as f:
    lines = f.readlines()
for i, line in enumerate(lines[:6]):
    rec = json.loads(line)
    print(rec['prompt'])
    display(Image(rec['final_path']))
```
